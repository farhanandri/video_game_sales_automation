Import library yang diperlukan

```python
import pandas as pd
from great_expectations.data_context import FileDataContext
```

Load data csv ke df

```python
df = pd.read_csv('vgsales.csv')
```

Melihat data

```python
df
```

Melihat informasi data

```python
df.info()
```

Mengecek duplicated data

```python
df.duplicated().sum()
```

Mengecek missing value data

```python
df.isna().sum()
```

Menghapus missing value dari kolom Year

```python
df.dropna(subset=['Year'], inplace=True)
```

Mengisi missing value dari kolom Publisher

```python
df["Publisher"] = df["Publisher"].fillna("Unknown")
```

Mengsortir ulang df berdasarkan kolom Global_Sales menurun dari yang terbesar

```python
df.sort_values(by='Global_Sales', ascending=False, inplace=True)
```

Mereset index rank dan membuat ulang rank agar urut dengan increment +1

```python
df.reset_index(drop=True, inplace=True)

df['Rank'] = df.index + 1
```

Mengecek missing value yang sudah dihandle

```python
df.isna().sum()
```

Melihat df

```python
df
```

Mengubah tipe data Year menjadi int

```python
df["Year"] = df["Year"].astype(int)
```

Mengecek informasi data

```python
df.info()
```

Menyimpan data bersih ke vgsales_clean

```python
df.to_csv('vgsales_clean.csv', index=False)
```

Membuat context directory dari GE

```python
context = FileDataContext.create(project_root_dir='./')
```

Membuat nama untuk datasource GE

```python
datasource_name = 'video_game_sales'
datasource = context.sources.add_pandas(datasource_name)
```

Menginisialisasi asset_name dan path

```python
asset_name = 'vgsales'
path_to_data = 'E:/Hacktiv8/M3/vgsales_clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)
batch_request = asset.build_batch_request()
```

Membuat expectation suite

```python
expectation_suite_name = 'expectation-video-game-sales'
context.add_or_update_expectation_suite(expectation_suite_name)
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

Validasi 1
Mengecek validasi value unique kolom Rank

```python
validator.expect_column_values_to_be_unique('Rank')
```

Validasi 2
Mengecek validasi value tidak boleh missing kolom Name

```python
validator.expect_column_values_to_not_be_null('Name')
```

Validasi 3
Mengecek validasi value kolom Year antara nilai min dan max kolom Year

```python
validator.expect_column_values_to_be_between(column='Year', min_value = df['Year'].min(), max_value = df['Year'].max())
```

Validasi 4
Mengecek validasi kolom harus ada pada kolom Na_Sales dan EU_ Sales

```python
validator.expect_column_to_exist(column='NA_Sales')
validator.expect_column_to_exist(column='EU_Sales')
```

Membuat variable placeholder untuk publisher

```python
publisher = df['Publisher'].unique().tolist()
```

Validasi 5
Mengecek validasi value Kolom Publisher harus ada pada set value yang telah ada sebelumnya pada variable publisher

```python
validator.expect_column_values_to_be_in_set('Publisher', value_set=publisher)
```

Validasi 6
Mengecek validasi tipe data float pada kolom JP_Sales dan Other_Sales

```python
validator.expect_column_values_to_be_in_type_list('JP_Sales', ['float'])
validator.expect_column_values_to_be_in_type_list('Other_Sales', ['float'])
```

Validasi 7
Mengecek validasi value menurut sesuai rank  pada kolom Global_Sales

```python
validator.expect_column_values_to_be_decreasing(column='Global_Sales')
```

Tugas Milestone 3 pre automasi selesai pada cell ini

cell dibawah opsional untuk menyimpan validator ke suite

```python
validator.save_expectation_suite(discard_failed_expectations=False)
```

membuat checkpoint

```python
checkpoint_1 = context.add_or_update_checkpoint(
    name = 'checkpoint_1',
    validator = validator,
)
```

run checkpoint

```python
checkpoint_result = checkpoint_1.run()
```

membuild docs dan path file docs

```python
context.build_data_docs()
```
